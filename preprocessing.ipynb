Importazione delle librerie necessarie

```python
import pandas as pd
import numpy as np
import os
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
```

Definizione dei percorsi dei file e delle variabili d'ambiente

```python
csv_extension = ".csv"
pickle_extension = ".pickle"

raw_folder_path = os.path.join('', 'raw')
data_folder_path = os.path.join('', 'data')

lifts_file_name = 'lifts'
runs_file_name = 'runs'
ski_areas_file_name = 'ski_areas'

lifts_raw_file_path = os.path.join(raw_folder_path, str(lifts_file_name) + str(csv_extension))
runs_raw_file_path = os.path.join(raw_folder_path, str(runs_file_name) + str(csv_extension))
ski_areas_raw_file_path = os.path.join(raw_folder_path, str(ski_areas_file_name) + str(csv_extension))

lifts_file_path = os.path.join(data_folder_path, str(lifts_file_name) + str(pickle_extension))
runs_file_path = os.path.join(data_folder_path, str(runs_file_name) + str(pickle_extension))
ski_areas_file_path = os.path.join(data_folder_path, str(ski_areas_file_name) + str(pickle_extension))

lifts = pd.read_csv(lifts_raw_file_path)
runs = pd.read_csv(runs_raw_file_path, low_memory=False)
ski_areas = pd.read_csv(ski_areas_raw_file_path)
```

```python
runs = runs.dropna(subset=["difficulty", "countries", "inclined_length_m", "ski_area_names"])
runs["difficulty"] = runs["difficulty"].str.lower()
runs["countries"] = runs["countries"].str.split(";")
runs = runs.explode("countries")
runs["ski_area_names"] = runs["ski_area_names"].str.split(";")
runs = runs.explode("ski_area_names")

ski_areas = ski_areas.dropna(subset=["countries", "lift_count", "downhill_distance_km", "nordic_distance_km"])
ski_areas["countries"] = ski_areas["countries"].str.split(";")
ski_areas = ski_areas.explode("countries")

lifts = lifts.dropna(subset=["lift_type"])
lifts["countries"] = lifts["countries"].str.split(";")
lifts = lifts.explode("countries")
```

```python
lifts.to_pickle(lifts_file_path)
runs.to_pickle(runs_file_path)
ski_areas.to_pickle(ski_areas_file_path)
```
